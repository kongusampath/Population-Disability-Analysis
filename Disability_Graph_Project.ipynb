# Project Description:
This project analyses data of people with disabilities in India, downloaded from data.gov.in in excel format. The dataset used in this project is from the year 2011. The data from the  excel sheet is captured in a Data Frame which is later used to generate a bar graph based on user input. Based on user input of State , Age-Group, Disability and Commmunity a bar graph is plotted depending on how columns the user wants to compare.

This project  uses Pandas which is Python for data analysis.
The graph is generated using matplotlib.
This project has been completed using ipython notebook.



```python
__author__ = 'KonguS'

#List of Modules
import numpy as np  #Numerical Pyton
import pandas as pd #Python for Data Analysis
import re #Regular Expressions
import matplotlib.pyplot as plt #Plotting of Graphs
%matplotlib inline 
#To Plot Graph Inline
```

```python
#Function to open excel file  and create a data frame
def fileopen():
     myfile = "C:/Users/KonguS/python/Projects/India_Disability_Data.xls"
     df = pd.read_excel(myfile)#Data Frame created

     df.Disability=df.Disability.replace('Total','Whole')#Replacing the word "Total with the word "Whole"
     print("Data Frame:")
     print("Number of rows = %d\n" %df.shape[0],"Number of columns = %d\n" %df.shape[1])#Prints the size of the original Data Frame
     return df
```

```python
#Function that drops columns that are not necessary for Data Frame
def coldrop(df):

    for i in df.columns[10:23]:
        df.drop(i, axis=1, inplace=True)
        
    df.drop('Table Name', axis=1, inplace=True)
    df.drop('District Code', axis=1, inplace=True)
    df.columns

   
```

```python
#Regular Expressions to convert column "Area Name" to a particular format
def regex(df):
    for i in df.index:
    
        df.loc[i,'Area Name']=re.sub(r'\s\([\d]*\)',"",df.loc[i,"Area Name"])
        df.loc[i,'Area Name']=re.sub(r'State - ' ,"",df.loc[i,"Area Name"])
```

```python
#Function to create a dictionary with all column names as key and a list of contents of each column and  
def lister(df):
    a=[[],[],[],[],[]] #List of five lists to be inserted as the value for the dictionary d
    d=dict()
    x=0
    for j in df.columns[0:5]:
       
        for i in (df.index):
            if(df.loc[i,j] not in a[x]):
                a[x].append(df.loc[i,j])
                d[j]=a[x]
        x+=1
    #Adding the column names to the key "Disabled Population" in the dictionary   
    d["Disabled Population"] = ['Total disabled population - Persons(Male+Female)','Total disabled population - Males','Total disabled population - Females']   
    return d
        
```
